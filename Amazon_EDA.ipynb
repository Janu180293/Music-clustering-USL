## Distribution of Audio Features

```python
import matplotlib.pyplot as plt

audio_features = [
    "danceability","energy","speechiness",
    "acousticness","instrumentalness",
    "liveness","valence","tempo","loudness"
]

df[audio_features].hist(
    figsize=(15,10),
    bins=30
)

plt.tight_layout()
plt.show()
```

## Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

sns.heatmap(
    df[audio_features].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.show()
```

## Outlier detection

```python
plt.figure(figsize=(15,8))

for i,col in enumerate(audio_features):
    plt.subplot(3,3,i+1)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

## Popularity distribution

```python
sns.histplot(
    df["popularity_songs"],
    bins=30
)

plt.show()
```

## Energy vs Danceability

```python
sns.scatterplot(
    x="danceability",
    y="energy",
    data=df,
    alpha=0.3
)

plt.show()
```
